```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ("canada_per_capita_income.csv")
data
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('salary')
plt.scatter(data.year,data.salary,color='red',marker='+')
```

```python
new_d = data.drop('salary',axis='columns')
new_d
```

```python
salary = data.salary
salary
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_d,salary)
```

```python
reg.predict([[2020]])
```
